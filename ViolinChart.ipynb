```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
```

```python
d = pd.read_csv('altcoins.csv')
d.head()
```

```python
d.fillna(0, inplace=True)
d.head()
```

```python
d.tail(5)
```

```python
data = d.tail(30)
# Choose the two columns and make a list
l = [data.ETH, data.ZEC, data.DASH, data.LTC, data.BTC]
# Instantiate a figure and axes
fig1, ax1 = plt.subplots(figsize = (20, 8))
bp = ax1.violinplot(
    l,
    vert = False,
    widths = [0.8, 0.8, 0.8, 0.8, 0.8],
    showmeans = True, 
    showmedians = True
)

ax1.set_title('BTC VS ALTS', size = 20)
ax1.set_yticks([1,2,3,4,5])
ax1.set_yticklabels(['ETH','ZEC','DASH','LTC','BTC'])
mft = StrMethodFormatter("${x}")

ax1.set_xscale('log')
##ax1.set_xticklabels(labels = "",fontsize=8)
ax1.xaxis.set_major_formatter(mft)
ax1.set_facecolor('#fff')
fig1.set_facecolor('#fff')

# Set the color of the violin patches
colors = ['Blue', 'Red', 'Purple','Yellow','Orange']
for pc, color in zip(bp['bodies'], colors):
    pc.set_facecolor(color)
# Set the color of the median lines
bp['cmedians'].set_colors(colors)
```
