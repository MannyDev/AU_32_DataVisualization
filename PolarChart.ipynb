```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
df = pd.read_csv('barchart-dv.csv')
fig, ax = plt.subplots(figsize = (12, 5), subplot_kw = dict(polar = True))
x = df.Age
y1 = df.Women
y2 = df.Men
# Women on top.. of the stacked bar chart
ax.bar(x, y1 + y2, label = 'Women', alpha = 0.8)

# Men below
ax.bar(x, y2, label = 'Men', alpha = 0.8)

# Remove Y-axis labels
ax.set_yticklabels([])

# Let's remove Y-axis ticks
ax.tick_params(axis='y', labelbottom='False')

l = ax.legend(loc = 'upper right')
l.set_bbox_to_anchor((0.3, 0.9))
```
