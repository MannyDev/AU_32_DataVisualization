```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
```

```python
d = pd.read_csv('altcoins.csv')
d.head()
```

```python
d.fillna(0, inplace=True)
d.head()
```

```python
d.tail(5)
```

```python
data = d.tail(30)
# Choose the two columns and make a list
l = [data.ETH, data.ZEC, data.DASH, data.LTC, data.BTC]
# Instantiate a figure and axes
fig1, ax1 = plt.subplots(figsize = (20, 8))
bp = ax1.boxplot(
    l,
    vert = False,
    whis = 1.5,
    widths = [0.8, 0.8, 0.8, 0.8, 0.8],
    labels = ['ETH', 'ZEC', 'DASH', 'LTC', 'BTC'],
    showmeans = True,
    meanprops = dict(marker = '*'),
    showcaps = True,
    showfliers = True,
    whiskerprops = dict(marker = 'o', markersize = 10, markeredgecolor = 'r'),
    medianprops = dict(linestyle = 'dotted', linewidth = 3)
)

ax1.set_title('BTC (ORANGE) VS ALTS (GREEN)', size = 20)

mft = StrMethodFormatter("${x}")

ax1.set_xscale('log')
##ax1.set_xticklabels(labels = "",fontsize=8)
ax1.xaxis.set_major_formatter(mft)
bp['boxes'][1].set(
    color = 'c',
    lw = 2,
    ls = 'dashed')

bp['fliers'][0].set(markersize = 5)
```
